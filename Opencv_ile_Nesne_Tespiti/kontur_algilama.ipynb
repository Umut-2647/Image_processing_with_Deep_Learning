```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("media/sekil.png",0)
plt.figure(), plt.imshow(img , cmap="gray"), plt.axis("off")
```

```python
contours ,hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE) #RETR_CCOMP: iç ve dış konturları bulur. CHAIN_APPROX_SIMPLE: konturları sadeleştirir.

external_contours = np.zeros(img.shape) #siyah bir resim oluşturduk
internal_contours = np.zeros(img.shape) 

for i in range(len(contours)): #her bir kontur için
    #external contours
    if hierarchy[0][i][3] == -1: #dış konturların indexi -1
        cv2.drawContours(external_contours,contours,i,255,-1)
    else: #iç konturlar
        cv2.drawContours(internal_contours,contours,i,255,-1)

plt.figure() , plt.imshow(external_contours,cmap="gray"), plt.axis("off")
```
