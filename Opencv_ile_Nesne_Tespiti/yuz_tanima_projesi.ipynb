```python
import cv2
import matplotlib.pyplot as plt
```

```python
einstein = cv2.imread("media/einstein.jpg", 0)
plt.figure() , plt.imshow(einstein, cmap="gray"), plt.axis("off"), plt.title("Einstein")
```

```python
#siniflandirici
face_cascade =cv2.CascadeClassifier("media/haarcascade_frontalface_default.xml")
face_rect = face_cascade.detectMultiScale(einstein) #yuz tespiti

for (x,y,w,h) in face_rect:
    cv2.rectangle(einstein,(x,y),(x+w,y+h),(255,255,255),10)

plt.figure() , plt.imshow(einstein, cmap="gray"), plt.axis("off"), plt.title("Einstein with face detection")
```

```python
#barcelona ornegi
barcelona = cv2.imread("media/barcelona.jpg", 0)

face_rect = face_cascade.detectMultiScale(barcelona,minNeighbors=3) #yuz tespiti

for (x,y,w,h) in face_rect:
    cv2.rectangle(barcelona,(x,y),(x+w,y+h),(255,255,255),3)

plt.figure() , plt.imshow(barcelona, cmap="gray"), plt.axis("off"), plt.title("Barcelona with face detection")
```

```python
#video ornegi
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if ret == False:
        break
    else :
        face_rect = face_cascade.detectMultiScale(frame,minNeighbors=3) #yuz tespiti

        for (x,y,w,h) in face_rect:

            cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,255),3)
            
        cv2.imshow("Face detection", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
