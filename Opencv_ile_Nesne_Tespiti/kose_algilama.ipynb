```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("media/sudoku.png",0)
img = np.float32(img) #float ile ondalikli sayilara cevirdik
print(img.shape)
plt.figure(),plt.imshow(img, cmap = "gray"),plt.axis("off"),plt.title("Original Image"),plt.show()
```

# HARRIS CORNER DETECTION

```python
dst = cv2.cornerHarris(img , blockSize=2, ksize=3 ,k=0.04) #ksize kenar tespiti icin kullanilan kenar buyuklugu , k ise kenar tespiti icin kullanilan bir sabit 
plt.figure(),plt.imshow(dst, cmap = "gray"),plt.axis("off"),plt.title("Corner Image"),plt.show()
```

```python
#tespit ettigi koseleri belirginlestirelim
dst = cv2.dilate(dst,None)
img[dst>0.2*dst.max()]= 1 #koseleri beyaz yapalim

plt.figure(),plt.imshow(dst, cmap = "gray"),plt.axis("off"),plt.show()
```

# SHI TAMSAI DETECTION

#GOOD FEATURES TO TRACK

```python
img = cv2.imread("media/sudoku.png",0)
img = np.float32(img) #float ile ondalikli sayilara cevirdik
```

```python
corners = cv2.goodFeaturesToTrack(img,100,0.01,10) #100 tane koseleri tespit et , 0.01 minumum mesafe , 10 minumum koseler arasi mesafe
corners = np.int64(corners) #koseleri integer a cevirdik

for i in corners:
    x,y = i.ravel() 
    cv2.circle(img,(x,y),3,(125,125,125),-1) #koseleri cember icine alalim

plt.imshow(img), plt.axis("off")
```
