```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('media/sudoku.png', 0)
plt.figure(),plt.imshow(img,cmap="gray"),plt.axis("off"),plt.title("Original Image"),plt.show()
```

```python
##x gradyanlari  (dik olan kenarlari tespit eder)

sobelx = cv2.Sobel(img,ddepth=cv2.CV_16S,dx=1,dy=0,ksize=5)
plt.figure(),plt.imshow(sobelx,cmap="gray"),plt.axis("off"),plt.title("Sobel X"),plt.show()
```

```python
##y gradyanlari  (yatay olan kenarlari tespit eder)

sobely = cv2.Sobel(img,ddepth=cv2.CV_16S,dx=0,dy=1,ksize=5)
plt.figure(),plt.imshow(sobely,cmap="gray"),plt.axis("off"),plt.title("Sobel Y"),plt.show()
```

```python
##laplacian gradyanlari  (hem x hem y gradyanlarini tespit eder)

laplacian = cv2.Laplacian(img,cv2.CV_16S)
plt.figure(),plt.imshow(laplacian,cmap="gray"),plt.axis("off"),plt.title("Laplacian"),plt.show()
```
