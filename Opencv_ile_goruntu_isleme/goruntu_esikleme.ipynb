```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("media/manzara_constrat.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #resmi siyah beyaza cevirir

#siyahlar 0 beyazlar 255


plt.figure() #yeni bir figure olusturur
plt.imshow(img,cmap="gray") #resmi gri gosterir
plt.axis("off") #eksenleri kapatir
plt.show() 
```

```python
##esikleme 

#60 degerin altindaki pikselleri 0 yapar ustundekileri 255 yapar

#THRESH_BINARY_INV tersini alir
_,thresh = cv2.threshold(img, thresh=60 , maxval=255 ,type=cv2.THRESH_BINARY) 

plt.figure() #yeni bir figure olusturur
plt.imshow(thresh, cmap="gray") #resmi gri gosterir
plt.axis("off") #eksenleri kapatir
plt.show()
```

```python
#uyarlamali esik degeri belirleme

thresh_img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY,11,8) #11 komsuluk 8 esik degeri

plt.figure() #yeni bir figure olusturur
plt.imshow(thresh_img, cmap="gray") #resmi gri gosterir
plt.axis("off") #eksenleri kapatir
plt.show()
```
