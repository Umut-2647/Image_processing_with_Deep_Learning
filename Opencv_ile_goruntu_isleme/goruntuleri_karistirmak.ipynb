```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('media/manzara1.jpg')
img1= cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) #renkleri RGB olarak gösterir

img2 = cv2.imread('media/manzara2.jpg')
img2= cv2.cvtColor(img2, cv2.COLOR_BGR2RGB) #renkleri RGB olarak gösterir



plt.figure() #renkleri RGB olarak gösterir
plt.imshow(img1)

plt.figure()
plt.imshow(img2)
```

```python

img1 = cv2.resize(img1, (600,600))
img2 = cv2.resize(img2, (600,600))

print(img1.shape)
print(img2.shape)



plt.figure() #renkleri RGB olarak gösterir
plt.imshow(img1)

plt.figure()
plt.imshow(img2)
```

```python
## karistirilmis resim = alpha*img1 + beta*img2

blended = cv2.addWeighted(src1=img1, alpha=0.5 , src2=img2 , beta=0.5, gamma=0) #alpha ve beta degerleri toplami 1 olmali

plt.figure()
plt.imshow(blended)
```
