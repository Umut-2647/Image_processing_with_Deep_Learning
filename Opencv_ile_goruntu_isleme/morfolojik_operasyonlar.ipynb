```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("media/team.png",0)

plt.figure(),plt.imshow(img,cmap="gray"),plt.axis("off"),plt.title("Original Image"),plt.show()
```

## EROZYON

```python
#kuculme islemi yapiyoruz

kernel= np.ones((5,5),dtype=np.uint8)

result = cv2.erode(img,kernel,iterations=2)
plt.figure(),plt.imshow(result,cmap="gray"),plt.axis("off"),plt.title("Erozyon Image"),plt.show()
```

## DILATÄ°ON (GENISLEME)

```python
result2 = cv2.dilate(img,kernel,iterations=1)
plt.figure(),plt.imshow(result2,cmap="gray"),plt.axis("off"),plt.title("Dilation Image"),plt.show()
```

```python
##white noise ekleyelim ve opening islemi yapalim

whiteNoise = np.random.randint(0,2,size=img.shape[:2]) #0 ve 1 arasinda random degerler uretir
whiteNoise = whiteNoise*255 # 0 ve 255 arasinda degerler olusturduk
plt.figure(),plt.imshow(whiteNoise,cmap="gray"),plt.axis("off"),plt.title("White Noise Image"),plt.show()
```

```python
noise_img = whiteNoise+ img
plt.figure(),plt.imshow(noise_img,cmap="gray"),plt.axis("off"),plt.title("Noised Image"),plt.show()
```

# ACÄ°LMA Ä°SLEMÄ° (WHÄ°TE NOISE GEREKLI)

```python

#beyaz gurultuyu temizlemek icin opening islemi yapalim
opening = cv2.morphologyEx(noise_img.astype(np.float32),cv2.MORPH_OPEN,kernel) #ondalikli sayiya cevirmeliyiz
plt.figure(),plt.imshow(opening,cmap="gray"),plt.axis("off"),plt.title("Opening Image"),plt.show()
```

```python
##black noise ekleyelim ve closing islemi yapalim
blackNoise = np.random.randint(0,2,size=img.shape[:2])
blackNoise = blackNoise*-255
plt.figure(),plt.imshow(blackNoise,cmap="gray"),plt.axis("off"),plt.title("Black Noise Image"),plt.show()
```

```python
blackNoise_img = img + blackNoise
blackNoise_img[blackNoise_img<=-245] =0
plt.figure(),plt.imshow(blackNoise_img,cmap="gray"),plt.axis("off"),plt.title("Black Noise Image"),plt.show()
```

## CLOSING (BLACK NOISE GEREKLI)

```python
#black noise temizlemek icin closing islemi yapalim

closing = cv2.morphologyEx(blackNoise_img.astype(np.float32),cv2.MORPH_CLOSE,kernel)
plt.figure(),plt.imshow(closing,cmap="gray"),plt.axis("off"),plt.title("Closing Image"),plt.show()
```

# GRADIENT 

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
plt.figure(),plt.imshow(gradient,cmap="gray"),plt.axis("off"),plt.title("Gradient Image"),plt.show()
```
