```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("media/red_blue.png")
img_vis = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # RGB ye Ã§evirme
plt.figure(), plt.imshow(img_vis)
```

```python
print(img.shape)
```

```python
img_hist = cv2.calcHist([img],[0],mask=None,histSize=[256],ranges=[0,256])
print(img_hist.shape)
plt.figure(), plt.plot(img_hist)
```

```python
color = ("b","g","r")
plt.figure()

for i,c in enumerate(color): # i = 0,1,2 c = b,g,r sirasiyla karsilikli birbirine esitler
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=c) #sirasyla b,g,r renklerini cizdirir
    
```

```python
golden_gate = cv2.imread("media/golden_gate.jpg")
golden_gate_vis = cv2.cvtColor(golden_gate,cv2.COLOR_BGR2RGB) # RGB ye Ã§evirme
plt.figure(), plt.imshow(golden_gate_vis)
```

```python
print(golden_gate.shape)
```

```python
mask = np.zeros(golden_gate.shape[:2],np.uint8) #kanallarla ilgilenmiyoruz
plt.figure(), plt.imshow(mask,cmap="gray")
```

```python
mask[400:800,200:700]=255 #hepsi beyaz olsun (yani ilgilendigimiz kisim disindaki kisimlar siyah olsun)
plt.figure(), plt.imshow(mask,cmap="gray")
```

```python
masked_Ä°mg_vis = cv2.bitwise_and(golden_gate_vis,golden_gate_vis,mask=mask) #maskeleme islemi
plt.figure(), plt.imshow(masked_Ä°mg_vis)
```

```python
masked_Ä°mg = cv2.bitwise_and(golden_gate,golden_gate,mask=mask) #maskeleme islemi
plt.figure(), plt.imshow(masked_Ä°mg)
```

```python
masked_Ä°mg_hst=cv2.calcHist([golden_gate],[0],mask=mask,histSize=[256],ranges=[0,256]) #0 kirmizi 1 yesil 2 mavi olur
plt.figure(), plt.plot(masked_Ä°mg_hst)
```
